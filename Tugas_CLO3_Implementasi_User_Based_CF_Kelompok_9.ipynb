```python
def make_prediction(a, b1, b2):
  ra = sum(a[0:19])/len(a[0:19])
  rb1 = sum(b1[0:20])/len(b1[0:20])
  rb2 = sum(b2[0:20])/len(b2[0:20])
  sim_ab1 = b1[18]
  sim_ab2 = b2[18]
  rb1p = b1[17]
  rb2p = b2[17]

  return ra + 1/(sim_ab1 + sim_ab2) * (sim_ab1 * (rb1p - rb1) + sim_ab2 * (rb2p - rb2 ))
```
